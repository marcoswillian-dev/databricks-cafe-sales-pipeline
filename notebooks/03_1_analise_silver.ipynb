```python
#Lendo arquivo da bronze para analise.
df = spark.table("`cafe_sales_2026`.bronze.dirty_cafe_sales")
display(df)
```

```python
%sql
--Conferindo a contagem de registros.
select count(*) from `cafe_sales_2026`.bronze.dirty_cafe_sales;
```

```python
#Utilizamos essa função para contar a quantidade de valores nulos em cada coluna do DataFrame, auxiliando na validação da qualidade dos dados antes das transformações.

from pyspark.sql.functions import col, when, sum

null_counts = df.select([
    sum(when(col(c).isNull(), 1).otherwise(0)).alias(c)
    for c in df.columns
])

display(null_counts)
```

```python
#Descobrir os valores existentes de cada coluna.
for c in df.columns:
    print(f"Coluna: {c}")
    df.groupBy(c) \
      .count() \
      .orderBy("count", ascending=False) \
      .show(10)
```
